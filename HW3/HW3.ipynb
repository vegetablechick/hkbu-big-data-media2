Such analysis bases on the titles and artists of their most favorable songs.

```python
pandas.read_csv('Works.csv')
```

```python
df = pandas.read_csv('Works.csv', header=None, names=['single', 'artist'])
```

```python
df['artist']
```

```python
all_artist=[]
for artist in df['artist']:
    all_artist.extend(str(artist).split())
```

```python
word_count={}
```

```python
for word in all_artist:
    if word not in word_count:
        word_count[word]=1
    else:
        word_count[word]=word_count[word]+1
```

```python
word_count
```

```python
df_word_count=pd.Series(word_count).to_frame().reset_index()
```

```python
df_word_count.sort_values(by=0,ascending=False)
```

```python
from wordcloud import WordCloud
wc=WordCloud
```

```python
w=wc().generate_from_frequencies(word_count)
```

```python
import matplotlib.pyplot as plt
plt.imshow(w)
plt.axis('off')
```
